```python
import pandas as pd

csv_file_path = './Caudales_norm_etiquetas_ajustado.csv'
df = pd.read_csv(csv_file_path)

df.head()
```

```python
valores_unicos_estacion = ['la_esperanza_nechi', 'la_raya', 'las_varas', 'magangue', 'sucre', 'tres_cruces', 'villa_teresa']
df = df.drop(columns=valores_unicos_estacion)
df.head()
```

```python
def transformar_valores(estacion):
    return estacion.lower().replace(' ', '_')

df['Estacion'] = df['Estacion'].apply(transformar_valores)

df['Estacion'].unique()
```

```python
valores_asociados = {
    'la_esperanza_nechi': {
        'Nombre Estacion': 'LA ESPERANZA NECHI [27037010]',
        'Area Operativa': 'Area Operativa 01 - Antioquia-Chocó'
    },
    'la_raya': {
        'Nombre Estacion': 'LA RAYA [25027910]',
        'Area Operativa': 'Area Operativa 01 - Antioquia-Chocó'
    },
    'las_varas': {
        'Nombre Estacion': 'LAS VARAS - AUT [25027200]',
        'Area Operativa': 'Area Operativa 01 - Antioquia-Chocó'
    },
    'magangue': {
        'Nombre Estacion': 'MAGANGUE [25027680]',
        'Area Operativa': 'Area Operativa 02 - Atlántico-Bolivar-Sucre'
    },
    'sucre': {
        'Nombre Estacion': 'SUCRE [25027110]',
        'Area Operativa': 'Area Operativa 02 - Atlántico-Bolivar-Sucre'
    },
    'tres_cruces': {
        'Nombre Estacion': 'TRES CRUCES [25027640]',
        'Area Operativa': 'Area Operativa 01 - Antioquia-Chocó'
    },
    'villa_teresa': {
        'Nombre Estacion': 'VILLA TERESA [25027860]',
        'Area Operativa': 'Area Operativa 02 - Atlántico-Bolivar-Sucre'
    }
}

# Agregar las nuevas columnas al dataframe
df['Nombre Estacion'] = df['Estacion'].apply(lambda x: valores_asociados[x]['Nombre Estacion'])
df['Area Operativa'] = df['Estacion'].apply(lambda x: valores_asociados[x]['Area Operativa'])
```

```python
estaciones = {}

for estacion in valores_unicos_estacion:
    estaciones[estacion] = df[df['Estacion'] == estacion]

estaciones.keys()
```

```python
estaciones["villa_teresa"].head()
```

```python
import os

output_dir = './Caudales_Parquet/'
os.makedirs(output_dir, exist_ok=True)

for estacion, dataframe in estaciones.items():
    filename = os.path.join(output_dir, f"{estacion}.parquet")
    dataframe.to_parquet(filename)

print("Archivos Parquet creados en './Caudales_Parquet/'.")
```
