```python
import pandas as pd
import os

# Ruta al archivo XLSB
xlsb_file_path = 'Historico EVA - desde el año 1987 al 2017.xlsb'

# Crear una carpeta para los CSV si no existe
output_folder = 'csv_exports'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Función para guardar cada hoja como CSV
def save_sheets_as_csv(xlsb_file, output_dir):
    with pd.ExcelFile(xlsb_file, engine='pyxlsb') as xls:
        for sheet_name in xls.sheet_names:
            print(f"Procesando hoja: {sheet_name}")
            df = pd.read_excel(xls, sheet_name=sheet_name)
            csv_file_path = os.path.join(output_dir, f"{sheet_name}.csv")
            df.to_csv(csv_file_path, index=False)
            print(f"Guardado: {csv_file_path}")

# Ejecutar la función
save_sheets_as_csv(xlsb_file_path, output_folder)
```
